## - Import Packages

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('~/Downloads/Terrorism/terrorism.csv', encoding ='ISO-8859-1', low_memory=False)
```

```python
df.head()   #displaying first five rows of the dataset|
```

```python
df.columns                                   #displaying the columnnames
```

```python
df.isnull().sum()                    #displaying the sum of null values in each column
```

```python
df2=df.copy()                      #copying the dataset into new dataframe
```

```python
df2.head()
```

```python

df2.dropna(axis=1,how='all')            #dropping the columns which has all null values
```

Renaming the columns so that the column name would be easy to understand

```python
df2.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

Considering only the useful columns

```python
df2=df2[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df2['Wounded'] = df2['Wounded'].fillna(0).astype(int)
df2['Killed'] = df2['Killed'].fillna(0).astype(int)     #filling the null value with zero
```

```python
df2['casualities']=df2['Killed']+df2['Wounded']            #adding a new column 'casuality'
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

## **Some basic analysis of the data set

```python
print("The country with highest terror attack:",df2['Country'].value_counts().index[0])
print("Regions with highest terrorist attacks: ",df2['Region'].value_counts().index[0])
print("Highest people killed in an attack",df2['Killed'].max(),"that too place in",df2.loc[df2['Killed'].idxmax()].Country)
print("City with the highest number of attacks:",df2['city'].value_counts().index[1])
print("Year with the most attacks:",df2['Year'].value_counts().idxmax())
print("Month with the most attacks:",df2['Month'].value_counts().idxmax())
print("Group with the most attacks:",df2['Group'].value_counts().index[1])
print("Most Attack Types:",df2['AttackType'].value_counts().idxmax())
```

```python
f,ax = plt.subplots(figsize=(8,8))
sns.heatmap(df2.corr(), annot=False, linewidths=.4, fmt =".1f=",cmap=sns.diverging_palette(20,220,n=200), ax=ax)
```

## The below plot shows the number of terrorist activities took place each year


```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df2,palette="coolwarm",edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title("Number of Terrorist Activity each Year")
plt.show()
```

## From the above plot you can say that in the year 2014 maximum terrorist activities took place

## The below plot shows the most used attacking method by the terrorists

```python
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=df2,palette='rocket',order=df2['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Attacking methods by terrorists")
plt.show()
```

## From the above plot we can say that Bombing/Explosion is the most used attacking method by terrorists

The below plot shows the favorite targets of terrorists

```python
plt.subplots(figsize=(15,6))
sns.countplot('Target_type',data=df2,palette='mako',order=df2['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Favorite Targets")
plt.show()
```

## From the above plot we can say private citizens and property are most targeted

The below plot shows the most affected region by terrorist activityÂ¶

The below plot shows the number of attacks and how much people are wounded and killed

```python
f,ax = plt.subplots(figsize=(18,10))
df2.Killed.plot(kind="line",color = "r", label="number of kills",linewidth=5, alpha=0.5, grid=True, linestyle=":")
df2.Wounded.plot(color="b", label="number of wounds",linewidth=5, alpha=0.5, grid=True,linestyle=":")
plt.legend(loc="upper right")
plt.xlabel("Number of attacks", size=15)
plt.ylabel("People", size=15)
plt.title("Line Plot")
```

## The below plot shows the number of attacks that took place and the no. of people killed

```python
coun_terror=df2['Country'].value_counts()[:15].to_frame()
coun_terror.columns=['Attacks']
coun_kill=df2.groupby('Country')['Killed'].sum().to_frame()
coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

## In the above plot we can see that maximum attacks and killings took place in iraq

The below plot shows the no. of killings that took place over the years and what type of weapon was most used

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
text = df2.Motive.dropna()
text = " ".join(str(motive) for motive in df2.Motive)
stopwords = set(STOPWORDS)
stopwords.update(["say","NaN","specific" ,"carried","incident","responsibility","claimed","noted","minority", "nothing",
                  "party","Party","noted","attack","motive","source","sources","stated","part","new", "us","The", "specific", "motive", "for",
                  "attack", "is", "unknown", "which", "Unknown","occurred", "occured","Occured", "state", "reported", "member", "group", "area", "related", "intended",
                  "larger","trend","may","target","targeted","suspected","speculated","retaliation","retalisation",
                  "says","call","unknown","nan","NAN","majority","communities","victim", "killed" ,"people", "posited",
                  "victims","attacks","response","United States","accused","iraq","note","stating",
                 "authorities","order","members"])
plt.subplots(figsize=(10,10))
wordcloud=WordCloud(stopwords=stopwords,background_color='black',width=512,height=382).generate(text)
plt.axis('off')
plt.title("What are the motives behind terrorism?")
plt.imshow(wordcloud)
plt.show()
```

## Therefore in the above wordcloud you can see what are the motives behind terrorism.
